```python
#copy 1 Exploration of Data visulization and utilization of Matplotlib Package and drawing plots
# it is usefull to gain insights and understand what happened in the past in a given context. it is also helpfull for feature engineering. 
#For implementing various plots in python we require the help of some liabries.
# 1.Matplotlib : Its a Python Plotting Libary mostly Used for data visulization. it makes use of numpy arrays, 
#                it is well intregrated with pandas to read columns and create plots
# 2.Seaborn: Its also a Python data visulation libary based on Matplotlib, it provide a high level interface for drawing innovative an
# informative statistical charts
# Q1. Draw a line in a diagram from position (1,3) to positon (8,10)
import matplotlib.pyplot as plt
import numpy as np

xpts=np.array([1,8])
ypts=np.array([3,10])

plt.plot(xpts,ypts)
plt.show()
```

```python
#BarChart
#its a frequency chart for qualitative(Catagorial) variables. It can be used to access the most occuring and
#the least occuring catagories within a data set.
#Question: Draw 4 bars.

xpts=np.array(["A","B","C","D"])
ypts=np.array([3,8,1,10])
plt.bar(xpts,ypts)
plt.show()
```

```python
#Question Draw 4 horizontally bar
xpts=np.array(["A","B","C","D"])
ypts=np.array([3,8,1,10])
plt.barh(xpts,ypts)
plt.show()
```

```python
#Barwidth
#Question: Draw 4 very thin bar
xpts=np.array(["A","B","C","D"])
ypts=np.array([3,8,1,10])
plt.bar(xpts,ypts,width=.2,color="cyan")
plt.show()
```

```python
#Note: for horizontal bars, the width changes to height
xpts=np.array(["india","america","china","brazil"])
ypts=np.array([30,80,100,105])
plt.barh(xpts,ypts,height=0.1)
plt.ylabel("Countries")
plt.xlabel("Population in Millions")
plt.title("Population of various countries")
plt.show()
```

```python
#Question: Draw Bar plot of a sample data frame.
import seaborn as sn
import pandas as pd
dataset=pd.DataFrame({'Country':['USA','Germany','Russia','Japan'],'Population':[1093,1166,1444,1226]})
sn.barplot(x="Country",y="Population",data=dataset,width=0.1)
plt.title('Population of the country')
plt.show()
```

```python
#Histogram
#It is a plot that shows the frequency distribution of a set of continious variable its gives an insight into the
#underline distribution of the variable outliers,sqewness etc.
#To draw a Histogram we utilize hist() method of matplotlib libary.
import pandas as pd
DF=pd.read_csv("CSBS_dataset1.csv")
```

```python
#Question: Draw the histogram for sold price in the IPL dataset and understand its distribution
plt.hist(DF['SOLD PRICE'],color="blue",edgecolor="black")
plt.title("Histogram for Sold Price")
plt.xlabel('SOLD PRICE')
plt.ylabel('Frequency')
plt.show()
#interpretation to be written, Histogram data shows the sold price is right skewed
#bydeafult it creates 10 pints in the histogram
```

```python
#
plt.hist(DF['SOLD PRICE'],color="red",edgecolor="black",bins=30)
plt.title("Histogram for Sold Price")
plt.xlabel('SOLD PRICE')
plt.ylabel('Frequency')
plt.show()
```

```python
#Distribution or Density PLot
#Its depicts the distribution of data over a continious interval. Its like a smooth histogramand visulises distribution
#of data
sn.distplot(DF['SOLD PRICE'],bins=30)
plt.title("Histogram for Sold Price")
plt.xlabel('SOLD PRICE')
plt.ylabel('Frequency')
plt.show()
```

```python
#Question(Write in copy) Draw the relationship between the feature , sixers and soldprice in the IPL dataset
Bats=DF[DF['PLAYING_ROLE']=='Batsman']
sn.regplot(x='SIXERS',y='SOLD PRICE',data=Bats)
plt.title('Scatter plot for SOLD PRICE versus the SIXERS with a regression line')
plt.xlabel('SIXERS')
plt.ylabel('SOLD PRICE')
plt.show()
```

```python
#PairPlot
#Take anytwo graphs, 2*2=4
#If there are many variable its not convinent to draw scatter plots, for each pair of variable to understand the relationship
#so a pair plot can be used to depict the relationship in a single diagram which can be plotted using pairplot() method
F=['SR_B','AVE','SIXERS','SOLD PRICE']
sn.pairplot(DF[F])
plt.show()
```

```python
#Corelation and Heatmap
#it is used to measuring the strength and measure the relationship between two continious random variable, its a statistical
#measure that in the extent  to which the variable change togeather
DF[F].corr()
```

```python
#A correlation heatmap is a graphical tool that displaces the corelatiom between multiple varible as a colour codded matrix
sn.heatmap(DF[F].corr(),annot=True)
plt.title("Heatmap of correlation values")
plt.show()
```
